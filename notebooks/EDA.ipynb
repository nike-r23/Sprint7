```python
import pandas as pd 
import plotly.express as px 

# Cargar los datos
df = pd.read_csv('../vehicles_us.csv')

# Ver las primeras filas
df.head()

# Informacin general del dataset
df.info()

#Estadisticas Basicas
df.describe()

#Crear histrograma de la condicion del vehiculo vs Ao del modelo
fig = px.histogram(df, x="model_year", color="condition", title="Condicin de Vehiculos por Ao del Modelo")
fig.show()

#Conocer el nombre de todas las columnas
print(df.columns.tolist())

#Realizar una comparativa de precios entre modelos 
#Saber cuantos modelos unicos tenemos# Ver los modelos nicos
print(f"Cantidad de modelos: {df['model'].nunique()} ")
top_models = df['model'].value_counts().index
filtered_data = df[df['model'].isin(top_models)]

#Histograma comparativo de precios por modelo 
fig = px.histogram(filtered_data, x="price", color="model", title="Distribucin de precios por modelo")
fig.show()
```
